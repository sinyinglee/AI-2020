```python
import requests
import re
import csv
import shutil
```

```python
from bs4 import BeautifulSoup
```

```python
letter_list = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]
```

```python
img_list = []
```

```python
height_list = []
web_list = []
```

```python
for i in letter_list:
    for num in range(1,37):
        web = f"https://celebsheight.org/celebrities-starting-with-{i}-heights?page={num}"
        page = requests.get(web)
        soup =  BeautifulSoup(page.content, "html.parser")
        img_links = soup.find_all("img", {"src": re.compile('.*?\.jpg')})
        for link in img_links:
            img_list.append(link["src"])
```

```python
for i in letter_list:
    for num in range(1,37):
        web= f"https://celebsheight.org/celebrities-starting-with-{i}-heights?page={num}"
        web_list.append(web)
```

```python
for web in web_list:
    page1 = requests.get(web)
    soup1 =  BeautifulSoup(page1.content, "html.parser")
    height = soup1.find_all("div", class_ = "celeb-grid-item-n")
    each_height = []
    for i in height:
         each_height.append(i.find("h2").text)
    height_list.append(each_height)
```

```python
emp_list = []
```

```python
for i in range(len(height_list)):
    for a in range(len(height_list[i])):
        emp_list.append(height_list[i][a])
    
```

### Save url into pnj.file and customize file name

```python
for i in range(len(img_list)):
    resp = requests.get(img_list[i], stream=True)
    local_file = open(emp_list[i],"wb")
    resp.raw.decode_content = True
    shutil.copyfileobj(resp.raw, local_file)
    del resp
```
